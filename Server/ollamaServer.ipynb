```python
import subprocess
import threading
import time
import os

# 1. Cấu hình để Ollama mở cửa đón khách
os.environ["OLLAMA_HOST"] = "0.0.0.0"
os.environ["OLLAMA_ORIGINS"] = "*"

# 2. Hàm khởi động server
def start_ollama_server():
    print("⏳ Đang khởi động Ollama Server...")
    # Lệnh này tương đương với việc gõ 'ollama serve' trong terminal
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# 3. Chạy server trong luồng riêng (để không bị treo máy)
thread = threading.Thread(target=start_ollama_server)
thread.start()

# 4. Đợi 20 giây cho server tỉnh ngủ hẳn
print("Đang đợi 20 giây cho server khởi động...")
time.sleep(20)

print("\n✅ SERVER ĐÃ BẬT! Giờ bạn hãy chạy lệnh curl ở dưới để kiểm tra.")
```

```python
import os
import shutil
!pip install pyngrok
from pyngrok import ngrok, conf

# --- DÁN CÁI MÃ BẠN VxỪA BẤM COPY VÀO GIỮA 2 DẤU NGOẶC KÉP ---
my_token = "PASTE_YOUR_TOKEN_HERE" 
# -----------------------------------------------------------

print("1. Đang tắt các kết nối cũ...")
ngrok.kill()

# Cắt bỏ khoảng trắng thừa cho chắc ăn
clean_token = my_token.strip()

print(f"2. Đang cài đặt token mới (bắt đầu bằng {clean_token[:5]}...)...")

# Xóa sạch cấu hình cũ để tránh lỗi
ngrok_path = os.path.join(os.path.expanduser("~"), ".ngrok2")
if os.path.exists(ngrok_path):
    shutil.rmtree(ngrok_path)

# Nạp token bằng lệnh hệ thống
os.system(f"ngrok config add-authtoken {clean_token}")

print("3. Đang mở kết nối...")
try:
    conf.get_default().config_path = None 
    
    # Kết nối cổng 11434
    public_url = ngrok.connect(11434).public_url
    
    print("\n" + "="*50)
    print("✅ THÀNH CÔNG RỒI! LINK CỦA BẠN LÀ:")
    print(public_url)
    print("="*50)
    
except Exception as e:
    print("\n❌ Vẫn lỗi: ", e)
```
